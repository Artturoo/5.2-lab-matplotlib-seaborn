#### We would like to have in one single plot the summary statistics of the feature `Age`. What kind of plot would you use? Plot it. 

```python
# your code here
sns.histplot(x=titanic.Age,
             bins=40,
             element='step');   


plt.axvline(titanic.Age.mean(),
            c='r',
            linewidth=3.,
            linestyle='--',
            label='media de edad')

plt.axvline(titanic.Age.median(),
            c='orange',
            linewidth=3.,
            linestyle='--',
            label='mediana de edad')

plt.axvline(24,
            c='black',
            linewidth=3.,
            linestyle='--',
            label='moda de edad')

plt.title('Edades Titanic')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.legend()
plt.show();
```

#### Now in addition to the summary statistics, we want to have in the same plot the distribution of `Age`. What kind of plot would you use? Plot it. 

```python
# your code here
sns.histplot(x=titanic.Age,
             bins=40,
             element='step',
             kde=True);   


plt.axvline(titanic.Age.mean(),
            c='r',
            linewidth=3.,
            linestyle='--',
            label='media de edad')

plt.axvline(titanic.Age.median(),
            c='orange',
            linewidth=3.,
            linestyle='--',
            label='mediana de edad')

plt.axvline(24,
            c='black',
            linewidth=3.,
            linestyle='--',
            label='moda de edad')

plt.title('Edades Titanic')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.legend()
plt.show();
```

#### Plot the correlation matrix using seaborn.

```python
# your code here
pd.plotting.scatter_matrix(titanic._get_numeric_data(), figsize=(15, 8));
```

```python
corr=titanic.corr(method='pearson')
```

```python
sns.set(style='white')   
mascara=np.triu(np.ones_like(corr, dtype=bool))   
cmap=sns.diverging_palette(0, 10, as_cmap=True)   


sns.heatmap(corr,
            mask=mascara,
            cmap=cmap,
            vmax=1,
            vmin=-1,
            center=0,
            square=True,
            linewidth=0.5,
            cbar_kws={'shrink': 0.5},
            annot=True
           );
```

#### What are the most correlated features?

```python
"""
Pclass y Fare
"""
```
